```python
##### without prediction losss 
import numpy as np
import matplotlib.pyplot as plt

size = 4
name_list = ["1","50","100","200"]
x = np.arange(size)
a = [1.02,1.24,1.24,1.22]
b = [1.03,1.22,1.23,1.21]
c = [1.04,1.53,1.50,1.41]

d = [1.02,1.09,1.09,1.07]

total_width, n = 0.8, 4
width = total_width / n
x = x - (total_width - width) / 2
plt.xlabel('Epoch Number')
plt.ylabel('NN Preserving')
plt.bar(x, a,  width=width, label='reference', color="#63b2ee")
plt.bar(x + width, d,  width=width, label='target',color="#76da91")
plt.bar(x + 2 * width, b, width=width, label='linear',tick_label=name_list,color="#f8cb7f")
plt.bar(x + 3 * width, c, width=width, label='autoencoder',color="#f89588")
plt.legend(loc=2, bbox_to_anchor=(1.05,1.0),borderaxespad = 0.)
plt.show()
```
